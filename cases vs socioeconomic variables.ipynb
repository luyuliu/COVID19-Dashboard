import library 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

prepare data

```python
case_state = pd.read_csv("corona.corona_cases_state_level.csv")
ses_state = pd.read_csv("ACSDP5Y2018.compiled_state.csv")
case_state_usa = case_state[case_state['Region'] == "USA"]
case_state_usa_agg = case_state_usa.groupby('State_Name').agg({"Confirmed": "max"})

state_data = pd.merge(ses_state, case_state_usa_agg, left_on = "State_Name", right_on = "State_Name")
state_data
print(state_data.columns)
```

set variable names

```python
pop = state_data['Tot_pop']
senior = state_data['Senior']
income = state_data['Med_HH_Inc']
pov = state_data['Pct_below_Poverty_Rate']
case = state_data['Confirmed']
print(state_data[state_data['Confirmed'] > 30000])
```

#Scatter Plot (pop vs case)

```python
plt.scatter(pop, case)
plt.xlabel("Population")
plt.ylabel("Total number of cases")
#plt.xlim(0,15000000)
plt.ylim(0,30000)
```

#Scatter Plot (senior vs case)

```python
plt.scatter(senior, case)
plt.xlabel("Senior")
plt.ylabel("Total number of cases")
#plt.xlim()
plt.ylim(0,30000)
```

#Scatter Plot (income vs case)

```python
plt.scatter(income, case)
plt.xlabel("Median Household Income")
plt.ylabel("Total number of cases")
#plt.xlim()
plt.ylim(0,30000)
```

#Scatter Plot (pct below poverty rate vs case)

```python
plt.scatter(pov, case)
plt.xlabel("Percentage below poverty rate")
plt.ylabel("Total number of cases")
plt.xlim(0,20)
plt.ylim(0,30000)
```
